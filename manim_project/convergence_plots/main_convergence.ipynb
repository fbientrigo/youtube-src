```python
import matplotlib.pyplot as plt
import numpy as np
from plot_sum import plot_summation

# Define the series you want to visualize
N = 20
```

```python
series_n = [n for n in range(N)]
plot_summation(series_n, r'$ S = \sum_{n=0}^\infty n $')
plt.show()
```

```python
series_2 = [1/(n*n) for n in range(1,N)]
plot_summation(series_2, r'$\sum_{n \geq 1} \frac{1}{n^2}$')
plt.show()
```

```python
# la serie harmonica, divergente
harmonic = [ 1/n for n in range(1,10*N) ]
fig, ax = plot_summation(harmonic, r'$\sum_n \frac{1}{n}$')
plt.show()
```

```python
# la utilizada en la exponencial
exp_sum = [ 1 / np.math.factorial(n) for n in range(N) ]
fig, ax = plot_summation(exp_sum, r'$ e^x = \sum_{n=0} \frac{x^n}{ n! } \rightarrow e^1 = \sum_{n=0} \frac{1}{n!} $')
ax.hlines(np.e, 0, N, color='black', label=f'$e = {np.e : .4f}$')
plt.legend()
plt.show()
```

```python
# la utilizada en la exponencial
exp_minus = [ (-1)**n / np.math.factorial(n) for n in range(N) ]
fig, ax = plot_summation(exp_minus, r'$ e^{-x} = \sum_{n=0} (-1)^n \frac{x^n}{ n! } \rightarrow e^{-1} = \sum_{n=0} \frac{(-1)^n}{n!} $')
ax.hlines(1/np.e, 0, N, color='black', label=f'$1/e = {1/np.e : .4f}$')
plt.legend()
plt.show()
```
